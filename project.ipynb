```python
import random
import matplotlib.pyplot as plt
```

```python
no_of_iterations=10000000  #1Crore
```

```python
# Simulate 100 coin flips and store the results in a list
results = []
for i in range(no_of_iterations):
    no_head=0
    for j in range(8):
        result = random.choice([0,1])
        if result:
            no_head = no_head+1
    results.append(no_head)
```

```python
frequency=[]
for i in range(8):
    frequency.append(results.count(i))
```

```python
print(frequency)
```

```python
# Plot the results on a bar graph
x = [0,1,2,3,4,5,6,7]
y = frequency
plt.bar(x, y)
plt.title('Results of '+str(no_of_iterations)+' Iterations')
plt.xlabel('No of Heads')
plt.ylabel('Frequency')
plt.plot(x,y, color='green')
plt.show()
```
